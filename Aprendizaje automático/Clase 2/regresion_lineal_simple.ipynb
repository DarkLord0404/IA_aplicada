## 1 - Introducción

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# descarga de datos
url_datos="https://raw.githubusercontent.com/NechuBM/dataset/main/datos_correlacion.txt"
!wget {url_datos} -q -O datos_x_y.txt

df_linea = pd.read_csv('datos_x_y.txt')
X = df_linea.X
Y = df_linea.Y
```

```python
plt.title('Diagrama de dispersion')
plt.xlabel('Variable de entrada')
plt.ylabel('Variable de salida')
plt.scatter(X, Y, c='green')
plt.show()
```

```python
def crear_linea(b0, b1):
    '''Crea una linea a partir de la pendiente y la intersección'''
    x_l = [0, 5]
    y_l = [b0 + b1*x for x in x_l]

    return x_l, y_l
```

```python
b0 = 7  # interseccion
b1 =  -2  # pendiente

x_l, y_l = crear_linea(b0, b1)


plt.figure(figsize=(10,6))
plt.title('Diagrama de dispersion')
plt.xlabel('Variable de entrada')
plt.ylabel('Variable de salida')
plt.scatter(X,Y)
plt.plot(x_l, y_l, c='red')
plt.show()
```
